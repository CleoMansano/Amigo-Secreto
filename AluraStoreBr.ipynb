#1. Análise do faturamento


```python
def formatar_faturamento(loja, numero_loja):
    faturamento = loja['Preço'].sum()
    faturamento_formatado = f"R$ {faturamento:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    print(f"Faturamento da Loja {numero_loja}: {faturamento_formatado}")
```

```python
lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    formatar_faturamento(l, i)
```

# 2. Vendas por Categoria


```python
def vendas_por_categoria(loja, numero_loja):
    vendas_categoria = loja.groupby('Categoria do Produto')['Preço'].sum().reset_index()
    vendas_categoria = vendas_categoria.sort_values(by='Preço', ascending=False)
    vendas_categoria.columns = ['Categoria', 'Total de Vendas']
    print(f"Vendas por Categoria - Loja {numero_loja}:\n{vendas_categoria}\n")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    vendas_por_categoria(l, i)
```

```python
def mostrar_venda_por_categoria(loja, numero_loja):
    venda_por_categoria = loja['Categoria do Produto'].value_counts()
    print(f"Vendas por categoria - Loja {numero_loja}:\n{venda_por_categoria}\n")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    mostrar_venda_por_categoria(l, i)
```

# 3. Média de Avaliação das Lojas

```python
def mostrar_media_avaliacao(loja, numero_loja):
    media = round(loja['Avaliação da compra'].mean(), 2)
    print(f"Média de Avaliação da Loja {numero_loja}: {media}")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    mostrar_media_avaliacao(l, i)
```

# 4. Produtos Mais e Menos Vendidos

```python
def mostrar_top_produtos(loja, numero_loja, top_n=5):
    top_produtos = loja['Produto'].value_counts().head(top_n).reset_index()
    top_produtos.columns = ['Produto', 'Quantidade']
    print(f"Top {top_n} produtos mais vendidos - Loja {numero_loja}:\n{top_produtos}\n")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    mostrar_top_produtos(l, i)
```

```python
def mostrar_produtos_menos_vendidos(loja, numero_loja, quantidade=5):
    menos_vendidos = loja['Produto'].value_counts(ascending=True).head(quantidade).reset_index()
    menos_vendidos.columns = ['Produto', 'Quantidade']
    print(f"{quantidade} Produtos menos vendidos - Loja {numero_loja}:\n{menos_vendidos}\n")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    mostrar_produtos_menos_vendidos(l, i)
```

# 5. Frete Médio por Loja


```python
def calcular_frete_medio(loja, numero_loja):
    frete_medio = round(loja['Frete'].mean(), 2)
    print(f"Frete médio da Loja {numero_loja}: R$ {frete_medio}")

lojas = [loja, loja2, loja3, loja4]

for i, l in enumerate(lojas, start=1):
    calcular_frete_medio(l, i)
```
